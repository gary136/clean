```python
def kind(x):
    if x=='None':
        return '無'
    elif x=='addedEstablishment':
        return '含增建或未登記建物'
    elif x=='expropriation':
        return '徵收用地'
    elif x=='relaTrade':
        return '特殊關係間之交易'
    elif x=='buyFromgov':
        return  '向政府機關承購'
    elif x=='notDecorated':
        return '毛胚屋'
    elif x=='allignConstruct':
        return '建商與地主合建'  
    elif x=='else':
        return '其他'
    elif x=='preSale':
        return '預售屋'  
def age(x):
    if x <= 5:
        return '5年以下'
    elif x > 5 and x <= 10:
        return '5~10年'
    elif x > 10 and x <= 15:
        return '10~15年'
    elif x > 15 and x <= 20:
        return '15~20年'
    elif x > 20 and x <= 25:
        return '20~25年'
    elif x > 25 and x <= 30:
        return '25~30年'
    elif x > 30 and x <= 35:
        return '30~35年'
    elif x > 35 and x <= 40:
        return '35~40年'
    elif x > 40:
        return '40年以上'
def transform(df):
    df['district'] = df['Address'].apply(lambda x:x[3:6])
    df['Address'] = df['Address'].apply(lambda x:x[6:])
    df.drop('geo',axis=1,inplace=True)
    df['floor'] = df['floor'].apply(lambda x:int(x))
    df['floor_sum'] = df['floor_sum'].apply(lambda x:int(x))
    df.drop('mrt',axis=1,inplace=True)
    df['special'] = df['special'].apply(kind)
    df['house_age'] = df['house_age'].apply(age)
    df = df[['Time', 'district', 'vil', 'Address', 'lat', 'long', 'Type', 'unit_price', 
            'house_age', 'special', 'house_area', 'land_area', 'car_place', 'neighbor', 'floor', 'floor_sum']]
    df.columns = ['dealTime', 'district', 'vil', 'address', 'lat', 'long', 'houseType', 'unitPrice', 
                  'houseAge', 'special', 'houseAarea', 'landArea', 'carPlace', 'neighbor', 'floor', 'floorSum']
    return df
df = transform(df)
```

```python
for name in data:
    print(name)
    df = pd.read_csv(name)
    df = transform(df)
#     if keyStatus == 0:
#         break
    df.to_csv(name, index=False)
```
